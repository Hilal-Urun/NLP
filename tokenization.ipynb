### Wath is a Tokenzier?

*Tokenization is breaking the raw text into small chunks. Tokenization breaks the raw text into words, sentences called tokens. These tokens help in understanding the context or developing the model for the NLP. The tokenization helps in interpreting the meaning of the text by analyzing the sequence of the words*

### A Minimal Tokenizer

Go [here](https://www.oreilly.com/library/view/applied-natural-language/9781492062561/ch04.html) for more examples

```python
text = open('/content/10000_4.txt', 'r').read()
words = text.split(" ")
tokens = {v: k for k, v in enumerate(words)}
```

```python
text
```

```python
tokens
```

```python
token_map = map(lambda t: tokens[t], words)
list(token_map)
```

NLTK also provides a simpler, regular-expression based tokenizer, which splits text on whitespace and punctuation:

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
text = open('/content/10000_4.txt', 'r').read()
wordpunct_tokenize(text)
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
sentences
```

```python
words = [word_tokenize(t) for t in sent_tokenize(text)]
words
```

### BPE tokenizer
Byte Pair Encoding, or BPE, is a subword segmentation algorithm that encodes rare and unknown words as sequences of subword units

Go [here](https://github.com/huggingface/tokenizers) to learn more about tokenizers library

For the following example we will use a subset of Amazon QA review dataset [1,2]. If you want to play with this dataset please go [here](http://jmcauley.ucsd.edu/data/amazon/qa/).

```python
from tokenizers import Tokenizer, AddedToken
from tokenizers.models import BPE, WordPiece

tokenizer = Tokenizer(
    BPE.from_file(
        vocab='/content/vocab.json',
        merges='/content/merges.txt',
        unk_token='<unk>')
)

output = tokenizer.encode("Hello, y'all! How are you 😁 ?")
print(output,"\n")

print(output.tokens)
```
